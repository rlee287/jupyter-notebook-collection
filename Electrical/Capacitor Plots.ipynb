```python
import numpy as np
import matplotlib.pyplot as plt
```

A normal and a semilog plot involving capacitor exponential decay.

$\tau= 8 \mu F * 390 \Omega = 0.00312 \, F\cdot\Omega$

$ V_0 = 5 \, V $

$$ V=V_0 e^{\frac{-t}{\tau}} = 5 e^{\frac{-t}{0.00312}} $$

Map $t$ from 0 to 0.1 and calculate $V$.

```python
t=np.linspace(0,0.1,100)
v=np.exp(-t/0.00312)
```

Plot a linear graph of voltage versus time.

```python
plt.plot(t,v)
```

```python
plt.show()
```

Plot a semilogy graph of voltage versus time.

```python
plt.semilogy(t,v)
```

```python
plt.show()
```

Simulate noise of amplitude $\pm0.1 mV$.

```python
v_noisy=v+(np.random.rand(*np.shape(t))*0.0002)-0.0001
```

```python
plt.plot(t,v,t,v_noisy)
```

```python
plt.show()
```

```python
plt.semilogy(t,v,t,v_noisy)
```

```python
plt.show()
```

```python
log_v_noisy=np.log(v_noisy)
slope_log=(log_v_noisy[2]-log_v_noisy[0])/(t[2]-t[0])
print(slope_log)
print("Tau should be "+str((-1)/slope_log))
```
