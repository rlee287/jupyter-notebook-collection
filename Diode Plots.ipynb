```python
import numpy as np
import matplotlib.pyplot as plt
```

A normal and a semilog plot involving Diode I-V curves.

$$ I=I_S\left(e^\frac{V_D}{n V_T}-1\right) $$

Map $t$ from 0 to 0.1 and calculate $V$.

```python
v=np.linspace(-1,1,200)
i=(10**-12)*np.exp((v/(1.3*0.026))-1)
```

Plot a linear graph of voltage versus time.

```python
plt.plot(v,i); plt.show()
```

Plot a semilogy graph of voltage versus time.

```python
plt.semilogy(v,i); plt.show()
```

Simulate noise of amplitude $\pm0.1 mV$.

```python
i_noisy=i+(np.random.rand(*np.shape(i))*0.0002)-0.0001
```

```python
plt.plot(v,i,v,i_noisy); plt.show()
```

```python
plt.semilogy(v,i,v,i_noisy); plt.show()
```

```python
log_i_noisy=np.log(i_noisy)
slope_log=(log_i_noisy[-1]-log_i_noisy[-4])/(v[-1]-v[4])
print("Exponent should be "+str(slope_log))
```
